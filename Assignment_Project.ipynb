```python
import streamlit as st
from PIL import Image
from transformers import ViltProcessor, ViltForQuestionAnswering
import torch


st.title(" Visual Question Answering App")


uploaded_image = st.file_uploader(" Upload an image", type=["jpg","jpeg","png"])

if uploaded_image is not None:
    image = Image.open(uploaded_image)
    st.image(image, caption="Uploaded Image", use_column_width=True)


question = st.text_input(" Ask a question about the image")


@st.cache_resource
def load_model():
    processor = ViltProcessor.from_pretrained("dandelin/vilt-b32-finetuned-vqa")
    model = ViltForQuestionAnswering.from_pretrained("dandelin/vilt-b32-finetuned-vqa")
    return processor, model

processor, model = load_model()


if uploaded_image is not None and question:
    st.write(" Thinking... please wait")


    inputs = processor(image, question, return_tensors="pt")

    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        predicted_idx = logits.argmax(-1).item()
        answer = model.config.id2label[predicted_idx]

    st.success(f" Answer: {answer}")
```

```python
import streamlit as st
import pdfplumber
import docx
import io
import json

st.title("📄 Resume Extractor (PDF & DOCX)")

# --- Upload ---
file = st.file_uploader("Upload your resume (.pdf or .docx)", type=["pdf", "docx"])


def read_text(uploaded_file):
    name = uploaded_file.name.lower()
    if name.endswith(".pdf"):
        text_parts = []
        with pdfplumber.open(uploaded_file) as pdf:
            for p in pdf.pages:
                t = p.extract_text()
                if t: text_parts.append(t)
        return "\n".join(text_parts)
    elif name.endswith(".docx"):
        d = docx.Document(uploaded_file)
        return "\n".join([p.text for p in d.paragraphs])
    return ""


EMAIL_RE = re.compile(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}")
PHONE_RE = re.compile(r"(\+?\d[\d\s\-()]{8,}\d)")


DEFAULT_SKILLS = {
    "python","java","c++","c","javascript","typescript","html","css","react","node",
    "django","flask","sql","mysql","postgres","mongodb","aws","azure","gcp",
    "git","docker","kubernetes","linux","bash","figma","wordpress","woocommerce",
    "ml","ai","pandas","numpy","opencv","nlp","streamlit"
}


with st.sidebar:
    st.header("⚙️ Options")
    custom_skills = st.text_area("Add custom skills (comma-separated)", "")
    extra = {s.strip().lower() for s in custom_skills.split(",") if s.strip()}
    skill_bank = DEFAULT_SKILLS.union(extra)
    show_preview = st.checkbox("Show text preview", True)

def extract_name(text):

    for line in text.splitlines():
        L = line.strip()
        if not L: continue
        if "@" in L: continue

        if len(L.split()) <= 6:
            return re.sub(r"[^A-Za-z .'-]", "", L).strip()
    return ""

def extract_email(text):
    m = EMAIL_RE.search(text)
    return m.group(0) if m else ""

def extract_phone(text):

    candidates = PHONE_RE.findall(text)
    if not candidates: return ""
    best = max(candidates, key=lambda x: len(re.sub(r"\D","",x)))
    return re.sub(r"\s+", " ", best).strip()

def extract_skills(text, bank):
    words = set(w.lower() for w in re.findall(r"[A-Za-z+#.]+", text))
    found = sorted([s for s in bank if s in words])
    return found

def extract_section(text, keywords):
    lines = text.splitlines()
    hit_idx = -1
    for i,l in enumerate(lines):
        L = l.lower()
        if any(k in L for k in keywords):
            hit_idx = i; break
    if hit_idx == -1: return ""
    block = lines[hit_idx: hit_idx+18]
    return "\n".join(block).strip()

def extract_education(text):
    return extract_section(text, ["education", "academics", "qualifications", "degree", "bachelor", "master", "bs", "ms"])

def extract_experience(text):
    return extract_section(text, ["experience", "employment", "work history", "professional experience", "internship"])


if file and st.button("Extract"):
    raw_text = read_text(file)
    if not raw_text.strip():
        st.error("No selectable text found (scanned PDF is not supported).")
        st.stop()

    text = re.sub(r"\r\n?", "\n", raw_text)
    text = re.sub(r"\t", " ", text)

    result = {
        "name": extract_name(text),
        "email": extract_email(text),
        "phone": extract_phone(text),
        "skills": extract_skills(text, skill_bank),
        "education": extract_education(text),
        "experience": extract_experience(text)
    }

    st.subheader("Extracted Fields")
    st.json(result)


    btn = st.download_button(
        label="Download JSON",
        data=json.dumps(result, indent=2),
        file_name="resume_extract.json",
        mime="application/json"
    )

    if show_preview:
        with st.expander(" Text Preview"):
            st.text(text[:2000] + ("..." if len(text) > 2000 else ""))

elif not file:
    st.info("Upload a PDF or DOCX resume to begin.")
```
