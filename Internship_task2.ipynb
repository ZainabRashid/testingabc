```python
import os
import shutil
import tempfile
import streamlit as st

from langchain_community.document_loaders import PyPDFLoader, TextLoader, Docx2txtLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline

from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationalRetrievalChain
from langchain.prompts import PromptTemplate


# -----------------------------
# UI: Title & Sidebar
# -----------------------------
st.set_page_config(page_title="RAG Chatbot", page_icon="💬", layout="wide")
st.title("💬 Conversational RAG Chatbot (LangChain + FAISS + HF)")

with st.sidebar:
    st.header("📚 Knowledge Base")
    uploaded_files = st.file_uploader(
        "Upload documents (PDF / TXT / DOCX)",
        type=["pdf", "txt", "docx"],
        accept_multiple_files=True,
    )
    build_btn = st.button("📦 Build / Rebuild Vector Store")
    st.divider()
    st.caption("Tip: You can upload multiple files together.")


# -----------------------------
# Paths & Helpers
# -----------------------------
DB_DIR = "./db_faiss"
os.makedirs(DB_DIR, exist_ok=True)

def load_docs_from_uploads(files):
    docs = []
    for uf in files:
        # save to temp file because loaders expect file paths
        with tempfile.NamedTemporaryFile(delete=False) as tmp:
            tmp.write(uf.read())
            tmp_path = tmp.name

        name = uf.name.lower()
        if name.endswith(".pdf"):
            loader = PyPDFLoader(tmp_path)
        elif name.endswith(".txt"):
            loader = TextLoader(tmp_path, encoding="utf-8")
        elif name.endswith(".docx"):
            loader = Docx2txtLoader(tmp_path)
        else:
            continue
        docs.extend(loader.load())
    return docs

def build_vectorstore(files):
    if not files:
        return None

    # 1) load
    raw_docs = load_docs_from_uploads(files)

    # 2) split
    splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=120)
    chunks = splitter.split_documents(raw_docs)

    # 3) embed + index
    embed = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    vectordb = FAISS.from_documents(chunks, embed)

    # 4) persist
    if os.path.exists(DB_DIR):
        shutil.rmtree(DB_DIR)
    vectordb.save_local(DB_DIR)
    return vectordb

def load_vectorstore_if_exists():
    if not os.path.isdir(DB_DIR):
        return None
    embed = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    try:
        return FAISS.load_local(DB_DIR, embed, allow_dangerous_deserialization=True)
    except Exception:
        return None


# -----------------------------
# Build / Load Vector Store
# -----------------------------
if build_btn:
    with st.spinner("Building vector store..."):
        vs = build_vectorstore(uploaded_files)
    if vs:
        st.success("Vector store built & saved ✅")
    else:
        st.error("No files processed. Please upload valid documents.")

vectordb = load_vectorstore_if_exists()
if not vectordb:
    st.info("No vector store found yet. Upload files and click **Build / Rebuild Vector Store**.")


# -----------------------------
# LLM (Hugging Face, no API key)
# -----------------------------
# A light, free text2text model — good enough for demos.
gen_pipe = pipeline(
    "text2text-generation",
    model="google/flan-t5-base",
    max_new_tokens=256,
    temperature=0.2,
)
llm = HuggingFacePipeline(pipeline=gen_pipe)

# -----------------------------
# Memory + Prompt + Chain
# -----------------------------
if "memory" not in st.session_state:
    st.session_state.memory = ConversationBufferMemory(
        memory_key="chat_history", return_messages=True
    )

RAG_PROMPT = PromptTemplate.from_template(
    """You are a helpful assistant. Use the retrieved context to answer the user's question.
If the answer isn't in the context, say you don't know. Be concise but complete.
Cite short quotes or section hints when useful.

Context:
{context}

Chat History:
{chat_history}

Question:
{question}

Answer:"""
)

def build_chain(vdb):
    if vdb is None:
        return None
    retriever = vdb.as_retriever(search_kwargs={"k": 4})
    chain = ConversationalRetrievalChain.from_llm(
        llm=llm,
        retriever=retriever,
        memory=st.session_state.memory,
        combine_docs_chain_kwargs={"prompt": RAG_PROMPT},
        verbose=False,
    )
    return chain

chain = build_chain(vectordb)

# -----------------------------
# Chat UI
# -----------------------------
st.subheader("Chat")

# initialize chat history UI store
if "messages" not in st.session_state:
    st.session_state.messages = []

# render prior messages
for m in st.session_state.messages:
    with st.chat_message(m["role"]):
        st.markdown(m["content"])

user_msg = st.chat_input("Ask anything about your uploaded documents…")
if user_msg and chain:
    # show user message
    st.session_state.messages.append({"role": "user", "content": user_msg})
    with st.chat_message("user"):
        st.markdown(user_msg)

    # run RAG
    with st.chat_message("assistant"):
        with st.spinner("Thinking…"):
            out = chain({"question": user_msg})
            answer = out["answer"] if isinstance(out, dict) and "answer" in out else str(out)
            st.markdown(answer)
            st.session_state.messages.append({"role": "assistant", "content": answer})

elif user_msg and not chain:
    st.warning("Please build the vector store first (upload files → Build).")
```

```python
tickets = [
    "My internet is down for the last 2 hours.",
    "I was double charged for my subscription this month.",
    "How do I reset my password?",
    "The app crashes whenever I upload a file.",
    "I want to upgrade my plan from basic to premium."
]

possible_tags = ["Connectivity", "Billing", "Account", "Technical Issue", "Upgrade"]
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

for ticket in tickets:
    result = classifier(ticket, candidate_labels=possible_tags, multi_label=True)
    print(f"\nTicket: {ticket}")
    print("Top 3 predicted tags:", result["labels"][:3])
```
